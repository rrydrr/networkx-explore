#**Import the libraries**

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

#**Load CSV Graph file**

```python
df = pd.read_csv('dataset/lastfm_asia_edges.csv')
G = nx.from_pandas_edgelist(df, 'node_1', 'node_2')
```

#**Load CSV Graph file**

```python
plt.figure(figsize=(10, 8))

nx.draw(G, with_labels=False, node_size=50, font_size=8)

plt.title('Graph from 5% of the Last.fm Asia Edges Dataset')

plt.show()
```
