#**Import the libraries**

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

**Load CSV Graph file**

```python
df = pd.read_csv('dataset/fb-pages-tvshow.csv')
G = nx.from_pandas_edgelist(df, 'node_1', 'node_2')
```

**Displaying Graph with networkX and matplotlib**

```python
plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=False, node_size=50, font_size=8)
plt.title('Graph from FB Pages TV Show Dataset')
plt.show()
```

**Community Detection using Label Propagation**

```python
communities = nx.algorithms.community.label_propagation_communities(G)

community_map = {}
for community_id, community in enumerate(communities):
    for node in community:
        community_map[node] = community_id

nx.set_node_attributes(G, community_map, 'community')
```
